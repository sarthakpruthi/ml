```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
data=pd.read_csv("homeprices.csv")
```

```python
data
```

```python
plt.xlabel("area",size=60,color="blue")
plt.ylabel("price",size=60,color="blue")
plt.scatter(data["area"],data["price"],color="red",marker="+")
```

```python
lr=linear_model.LinearRegression()
```

```python
lr.fit(data[["area"]],data["price"])
```

```python
%matplotlib inline
plt.xlabel("area",size=60,color="blue")
plt.ylabel("price",size=60,color="blue")
plt.scatter(data["area"],data["price"],color="red",marker="+")
plt.plot(data[["area"]],lr.predict(data[["area"]]))
```

```python
df=pd.read_csv("areas.csv")
```

```python
df.head()
```

```python
lr.predict([[1000]])
```

```python
%matplotlib inline
plt.xlabel("area",size=60,color="blue")
plt.ylabel("price",size=60,color="blue")
plt.scatter(data[["area"]],lr.predict(data[["area"]]),color="blue")
plt.plot(df[["area"]],lr.predict(df[["area"]]),color="orange")
```


# pickle to read file

```python
import pickle
```

```python
with open("model","wb") as f:
    pickle.dump(lr,f)
```

```python
with open("model","rb") as f:
    m=pickle.load(f)
```

```python
m.predict([[4000]])
```

# joblib to read file faster than pickle

```python
from sklearn.externals import joblib
```

```python
joblib.dump(lr,"joblib_model")
```

```python
mp=joblib.load("joblib_model")
```

```python
mp.predict([[4000]])
```
