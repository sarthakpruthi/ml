```python
import pandas as pd
```

```python
data=pd.read_csv("test_scores.csv")
data
```

```python
a=list(zip(data.math,data.cs))
df=pd.DataFrame(data=a,columns=["x","y"])
df
```

```python
def gradient(x,y):
    m=c=0
    iterations=10000
    learning_rate=0.00021111
    n=len(x)
    for i in range (iterations):
        y_pred=m*x+c
        md=(-2/n)*sum(x*(y-y_pred))
        cd=(-2/n)*sum(y-y_pred)
        cost=(1/n)*sum([val**2 for val in (y-y_pred)])
        m=m-learning_rate*md
        c=c-learning_rate*cd
        print("m={}, c={}, cost={}, iteration={}".format(m,c,cost,i))
```

```python
import numpy as np
x=np.array([data.math])
y=np.array([data.cs])
x,y
```

```python
gradient(data["math"],data["cs"])
```
