```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
```

```python
company=['google','amazon','walmart','patanjali']
revenue=[97,68,73,80]
profit=[63,22,29,35]
```

```python
ypos=np.arange(len(company))
ypos
```

```python
plt.xticks(ypos,company)
plt.bar(ypos,revenue,label="revenue")
plt.bar(ypos,profit,label="profit")
plt.legend()
#plt.grid()
```

```python
plt.xticks(ypos,company)
plt.bar(ypos-0.2,revenue,width=0.4,label="revenue")
plt.bar(ypos+0.2,profit,width=0.4,label="profit")
plt.legend()
```

```python
plt.xticks(ypos,company)
plt.barh(ypos-0.2,revenue,height=0.4,label="revenue")
plt.barh(ypos+0.2,profit,height=0.4,label="profit")
plt.legend()
```

```python
blood_sugar=[113,85,90,150,149,88,93,115,135]
```

```python
plt.hist(blood_sugar,bins=[80,90,100,110,120,130],rwidth=0.95)
```

```python
blood_sugarmen=[113,85,90,150,149,88,93,115,135]
blood_sugarwomen=[103,75,80,140,139,78,83,105,125]
```

```python
plt.xlabel('sugar range')
plt.ylabel('total number of patients')
plt.hist([blood_sugarmen,blood_sugarwomen],bins=[80,100,125],label=['men','women'],orientation='horizontal')
plt.legend()
```
