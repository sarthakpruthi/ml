```python
import pandas as pd
import numpy as np
```

```python
x=np.array([1,2,3])
y=np.array([5,6,7])
x,y
```

```python
def gradient(x,y):
    m_curr=b_curr=0
    iteration=1000
    n=len(x)
    learning_rate=0.04
    for i in range (iteration):
        y_pred=m_curr*x+b_curr
        md=(-2/n)*sum(x*(y-y_pred))
        bd=(-2/n)*sum(y-y_pred)
        cost_func=(1/n)*sum([val**2 for val in (y-y_pred)])
        m_curr=m_curr-learning_rate*md
        b_curr=b_curr-learning_rate*bd
        print("m={} , b={}, iteration={} , cost={}".format(m_curr,b_curr,i,cost_func))
```

```python
gradient(x,y)
```
