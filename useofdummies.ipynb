```python
 import pandas as pd
```

```python
data=pd.read_csv("homeprices.csv")
data
```

```python
dummies=pd.get_dummies(data["town"])
```

```python
df=pd.concat([data,dummies],axis="columns")
df
```

```python
df.drop(["town","west windsor"],axis="columns",inplace=True)
df
```

```python
df.rename(columns={"monroe township": "t1","robinsville":"t2"},inplace=True)
df
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression() 
```

```python
lr.fit(df[["area","t1","t2"]],df[["price"]])
```

```python
lr.predict([[3200,1,0]])
```

```python
lr.score(df[["area","t1","t2"]],df[["price"]])
```
