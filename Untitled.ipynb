```python
data=pd.read_csv("canada_per_capita_income.csv")
data.head()
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(data[["year"]],data[["per capita income (US$)"]])
lr.score(data[["year"]],data[["per capita income (US$)"]])
lr.predict([[1990]])

def preprocess(data,n):
    df=data.copy()
    for i in (2,n):
        df["year{}".format(i)]=data["year"]**i
    return df

df=preprocess(data[["year"]],4)
lr.fit(df,data[["per capita income (US$)"]])
lr.score(df,data[["per capita income (US$)"]])
```

# gradient descent

example

```python
def gradientdescent(x,y):
    m=0
    c=0
    learning_rate=0.1802
    n=len(x)
    iter=1000
    for i in range(iter):
        y_pred=m*x+c
        cost=(1/n)*sum([i**2 for i in (y-y_pred)])
        dm=-(2/n)*sum(x*(y-y_pred))
        dc=-(2/n)*sum(y-y_pred)
        m=m-learning_rate*dm
        c=c-learning_rate*dc
        print("i={} , cost={},m={}".format(i,cost,m))
import numpy as np
gradientdescent(np.array([1,2,3]),np.array([4,5,6]))
```
